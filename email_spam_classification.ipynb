# **Email Spam Classification**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
df = pd.read_csv("emails.csv")
df.head(20)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df1=df.drop(columns=['Email No.'])

df1.corr()
```

```python
X = df.iloc[:,1:3001]
X
```

```python
Y = df.iloc[:,-1].values
Y
```

```python
train_x,test_x,train_y,test_y = train_test_split(X,Y,test_size = 0.25)
```

# Naive Bayes

```python
mnb = MultinomialNB(alpha=1.9)
mnb.fit(train_x,train_y)
y_pred1 = mnb.predict(test_x)
print("Accuracy Score for Naive Bayes : ", accuracy_score(y_pred1,test_y))
print("\n\n\n")
print(classification_report(test_y, y_pred1))
```

# Support Vector Machines

```python
svc = SVC(C=1.0,kernel='rbf',gamma='auto')

svc.fit(train_x,train_y)
y_pred2 = svc.predict(test_x)
print("Accuracy Score for SVC : ", accuracy_score(y_pred2,test_y))
print("\n\n\n")
print(classification_report(test_y, y_pred2))
```

# Random Forests (Bagging)(Ensemble Learning Model)


```python
rfc = RandomForestClassifier()
param_grid = {'n_estimators': [50, 100, 150], 'criterion': ['gini', 'entropy']}
grid_rfc = GridSearchCV(rfc, param_grid, cv=5)
grid_rfc.fit(train_x, train_y)
best_rfc = grid_rfc.best_estimator_
y_pred3 = best_rfc.predict(test_x)
print("Accuracy Score of Random Forest Classifier:", accuracy_score(y_pred3, test_y))
print("\n\n\n")
print(classification_report(test_y, y_pred3))
```

Random Forest Classifier performs the best among the three. Decision tree classifiers are excellent classifiers. Random forest is a popular ensemble model that uses a forest of decision trees. So, obviously, combibining the accuracy of 100 trees (as n_estimators=100 here), will create a powerful model.
